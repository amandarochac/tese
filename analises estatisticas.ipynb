```python
def calculaCV(ee,titulo):
    temp = ee.describe()
    var=['betaglicosidase', 'cmcase','fpase', 'xilanase']
    cv = {}
    mini = {}
    maxi = {}
    for valor in var:
        cv[valor] = (temp[:][valor]['std']/temp[:][valor]['mean'])*100
        mini[valor] = temp[:][valor]['min']
        maxi[valor] = temp[:][valor]['max']

    print(titulo,'\n\tEnzima','\t\tMin','\tMáx','\tCV:')
    for valor in var:
         print('%15s %10.2f %10.2f %10.2f'%(valor,mini[valor], maxi[valor], cv[valor]))
```

```python
calculaCV(celluclast,'celluclast')
```

```python
celic.describe()
```

```python
calculaCV(celic,'celic')
```

```python
eeta.describe()
```

```python
calculaCV(eeta,'EETA')
```
